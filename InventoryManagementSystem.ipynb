```python
import json
import time

print("\n*******************************************************************\n\t\t\tWelcome to Inventory Management System\n*******************************************************************\n")
print("Chose the Function you want to perform:\n1.ADD ITEMS TO THE INVENTORY\n2.PURCHASE ITEMS FROM THE INVENTORY")
userInput = int(input("Enter 1 or 2 to proceed: "))


if(userInput == 1):

    itemNumber = int(input("Enter the number of Items that you want to add: "))
    fd = open("record.json", 'r')
    r = fd.read()
    fd.close()
    record = json.loads(r)

    for i in range(itemNumber):
        pr_id = str(input("Enter Product ID: "))
        if (pr_id in record.keys()):
            print("Product Already Exists!")
            qn = int(input("Enter quantity to be added: "))
            if(qn>=0):
                record[pr_id]['qn'] = record[pr_id]['qn'] + qn
            else:
                print("Invald Quantity.")
        else:
            name = str(input("Enter name: "))
            pr = int(input("Enter price: "))
            qn = int(input("Enter quantity: "))
            record[pr_id] = {'name': name, 'pr': pr, 'qn': qn}
            print("New product added to inventory")
        js = json.dumps(record)
        fd = open("record.json", 'w')
        fd.write(js)
        fd.close()


elif(userInput == 2):

    print("\nINVENTORY:")
    fd = open("record.json", 'r')
    r = fd.read()
    fd.close()
    record = json.loads(r)
    price = 0
    keySale = 0
    for key, value in record.items():
        print(key,":", value['name'], "- RS:", value['pr'])
    itemNumber = int(input("Enter the number of Items that you want to Purchase: "))
    for i in range(itemNumber):
        iprice = 0
        ipr_id = str(input("Enter Product ID: "))
        pr_id = ipr_id
        iqn = int(input("Enter quantity: "))
        qn = iqn
        price = price + (record[pr_id]['pr'] *qn)
        iprice = (record[pr_id]['pr'] * qn)
        if (record[pr_id]['qn']>=qn):
            record[pr_id]['qn'] = record[pr_id]['qn'] - qn
        else:
            print("Invald Quantity.")
            exit()
        js = json.dumps(record)
        fd = open("record.json", 'w')
        fd.write(js)
        fd.close()

        keySale = keySale + 1
        itime = time.ctime()
        sales = {keySale : {'product' : ipr_id, 'quantity' : iqn, 'amount': iprice, 'time of purchase' : itime}}

        jsale = json.dumps(sales)
        fdsale = open("sales.json", 'w')
        fdsale.write(jsale)
        fdsale.close()

    print("Billing Amount: ", price)
```
